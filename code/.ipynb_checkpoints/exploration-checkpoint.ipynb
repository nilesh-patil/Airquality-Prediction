```python
%matplotlib inline

import pandas as pd
import numpy as np
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/particulates/hourly_88101_2015.csv')
```

```python
data.columns = [x.replace(' ','') for x in data.columns]
```

```python
data.columns
```

```python
q = data[['StateName','CountyName','MethodName','DateLocal','TimeLocal','SampleMeasurement']]
```

```python
counties = data.CountyName.unique()
```

```python
q = q[q.CountyName==counties[0]]
q = q[q.MethodName==q.MethodName.unique()[0]]
```

```python
q.head()
```

```python
q.shape
```

```python
pt = plt.figure(figsize=(15,10))
sns.set_style('whitegrid')


pt.add_subplot(221)

sns.set_style('whitegrid')
sns.tsplot(data = q['SampleMeasurement'],color='orange')
sns.plt.xlabel('Reading Number - 2015')
sns.plt.ylabel('PM2.5 level( $\mu$gm/$m^3$)')
sns.plt.title('PM2.5 Reading in Jefferson - 2015')


pt.add_subplot(223)

sns.kdeplot(q['SampleMeasurement'],color='skyblue')
sns.plt.xlabel('Density')
sns.plt.xlabel('PM2.5 level( $\mu$gm/$m^3$)')
sns.plt.title('PM2.5 Reading in Jefferson - 2015')



pt.add_subplot(222)

sns.set_style('whitegrid')
sns.tsplot(data = data[data.CountyName=='Bronx']['SampleMeasurement'],color='orange')
sns.plt.xlabel('Reading Number - 2015')
sns.plt.ylabel('PM2.5 level( $\mu$gm/$m^3$)')
sns.plt.title('PM2.5 Reading in Bronx - 2015')


pt.add_subplot(224)

sns.kdeplot(data[data.CountyName=='Bronx']['SampleMeasurement'],color='skyblue')
sns.plt.xlabel('Density')
sns.plt.xlabel('PM2.5 level( $\mu$gm/$m^3$)')
sns.plt.title('PM2.5 Reading in Bronx - 2015')


plt.savefig('JeffersonVSBronx2.png',dpi=300)
```

```python
del(data)
```
